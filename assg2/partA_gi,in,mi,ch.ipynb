```python
def square(val):
    return val*val;
import math;
def nos(c,in1):
    in2=np.where(c[in1[0]]==0)
    in3=np.where(c[in1[0]]>0)
    return len(in2[0]),len(in3[0])

def giniscal(f,c):
    mf=np.median(f)
    in1=np.where(f<=mf)
    in2=np.where(f>mf)
    n00,n01=nos(c,in1)
    n10,n11=nos(c,in2)
    if (n00==0 or n01==0):
        p00=0.5
        p01=0.5
    else:
        p00=n00/(n00+n01)
        p01=n01/(n00+n01)
    if(n10==0 or n11==0):
        p10=0.5
        p11=0.5
    else:
        p10=n10/(n10+n11)
        p11=n11/(n10+n11)
    gl=1-p00*p00-p01*p01;
    gr=1-p10*p10-p11*p11;
    fgs=len(in1[0])/(len(in1[0])+len(in2[0]))*gl+len(in2[0])/(len(in1[0])+len(in2[0]))*gr
    return fgs
def infoGain(f,c):
    mf=np.median(f)
    in1=np.where(f<=mf)
    in2=np.where(f>mf)
    n00,n01=nos(c,in1)
    n10,n11=nos(c,in2)
    en0=1
    en1=1
    if(n00==0 or n01==0):
        en0=0
    else:
        p00=n00/(n01+n00);
        p01=n01/(n01+n00);
    if(n10==0 or n11==0):
        en1=0
    else:
        p10=n10/(n10+n11)
        p11=n11/(n10+n11)
    if(en0!=0):
        en0=-1*((p00*(math.log(p00,2)))+(p01*(math.log(p01,2))))
    if(en1!=0):
        en1=-1*((p10*(math.log(p10,2)))+(p11*(math.log(p11,2))))
    val0=n00+n10
    val1=n01+n11
    total=val0+val1;
    part0=n00+n01;
    part1=n10+n11;
    pr0=val0/(val0+val1)
    pr1=val1/(val0+val1)
    enP=-1*((pr0*(math.log(pr0,2)))+(pr1*(math.log(pr1,2))))
    entr=enP-(part0*en0/total)-(part1*en1/total)
    return entr;
def gainRatio(f,c):
    mf=np.median(f)
    in1=np.where(f<=mf)
    in2=np.where(f>mf)
    n00,n01=nos(c,in1)
    n10,n11=nos(c,in2)
    gain=infoGain(f,c)
    part0=n00+n01
    part1=n10+n11
    total=part0+part1
    spart0=part0/total;
    spart1=part1/total;
    if(spart0==0):
        contr0=0
    else:
        contr0=math.log(spart0,2)*spart0
    if(spart1==0):
        contr1=0
    else:
        contr1=(math.log(spart1,2)*spart1)
        
    splitInfo=-1*(contr0+contr1);
    if(splitInfo==0):
        fans=1
    else:
        fans=gain/splitInfo
    return fans
    
def misclassificationError(f,c):
    mf=np.median(f)
    in1=np.where(f<=mf)
    in2=np.where(f>mf)
    n00,n01=nos(c,in1)
    n10,n11=nos(c,in2)
    val0=n00+n10;
    val1=n01+n11;
    pr0=val0/(val0+val1);
    pr1=val1/(val0+val1);
    fans=1-max(pr0,pr1);
    return fans;
def chiValueFun(f,c):
    mf=np.median(f)
    in1=np.where(f<=mf)
    in2=np.where(f>mf)
    n00,n01=nos(c,in1)
    n10,n11=nos(c,in2)
    c1y=n11
    c1n=n10
    c0y=n01
    c0n=n00
    valy=c1y+c0y;
    valn=c1n+c0n;
    val1=c1y+c1n;
    val0=c0y+c0n;
    summ=valy+valn+val1+val0;
    
    auxval1y=valy*val1/summ;
    auxval0y=valy*val0/summ;
    auxval1n=valn*val1/summ;
    auxval0n=valn*val0/summ;
    if(auxval1y==0):
        temp1y=0;
    else:
        temp1y=square(auxval1y-c1y)/auxval1y;
    if(auxval0y==0):
        temp0y=0;
    else:
        temp0y=square(auxval0y-c0y)/auxval0y;
    if(auxval1n==0):
        temp1n=0;
    else:
        temp1n=square(auxval1n-c1n)/auxval1n;
    if(auxval0n==0):
        temp0n=0;
    else:
        temp0n=square(auxval0n-c0n)/auxval0n;   
    
    
    
    return temp1y+temp0y+temp1y+temp0n;
    
```

```python
#rank the features
def sortandReturnTop(arr):
    dicti={}
    flist=list()
    falist=list()
    for i in range(0,len(arr)):
        falist.append(arr[i])
    for i in range(0,len(arr)):
        dicti[falist[i]]=i
    falist.sort(reverse=True);
    for i in range(0,5):
        flist.append(dicti[falist[i]])
    return flist
```
